```python
import os

# Definisikan path dataset Anda  
DATASET_PATH = r'D:\backup\download\house-prices-advanced-regression-techniques\SAWITPRO\count\dataset'  
TRAIN_IMAGES_PATH = os.path.join(DATASET_PATH, 'images/train')  # Gambar pelatihan  
VAL_IMAGES_PATH = os.path.join(DATASET_PATH, 'images/val')      # Gambar validasi  
TRAIN_LABELS_PATH = os.path.join(DATASET_PATH, 'labels/train')  # Label pelatihan  
VAL_LABELS_PATH = os.path.join(DATASET_PATH, 'labels/val')      # Label validasi  

def check_labels(image_dir, label_dir):
    images = sorted([f.split('.')[0] for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))])
    labels = sorted([f.split('.')[0] for f in os.listdir(label_dir) if f.endswith('.txt')])
    missing_labels = set(images) - set(labels)
    if missing_labels:
        print("Images without labels:", missing_labels)
    else:
        print("All images have corresponding labels!")

check_labels(TRAIN_IMAGES_PATH,
             TRAIN_LABELS_PATH)
check_labels(VAL_IMAGES_PATH,
             VAL_LABELS_PATH)
```

```python
import os

def check_empty_labels(label_dir):
    empty_files = []
    for label_file in os.listdir(label_dir):
        if label_file.endswith('.txt'):
            file_path = os.path.join(label_dir, label_file)
            if os.stat(file_path).st_size == 0:  # Cek apakah file kosong
                empty_files.append(label_file)
    return empty_files

empty_val_labels = check_empty_labels('D:/backup/download/house-prices-advanced-regression-techniques/SAWITPRO/dataset/labels/train')

print("Empty val labels:", empty_val_labels)
```

```python
# 6. Test model
import cv2
import numpy as np
import os
from ultralytics import YOLO

# Fungsi untuk sliding window dengan overlap
def sliding_window(image, window_size=(512, 512), step_size=256):
    """
    Fungsi untuk membagi gambar besar menjadi potongan-potongan kecil dengan ukuran tertentu
    dan overlap antar potongan.
    """
    (h, w) = image.shape[:2]
    for y in range(0, h - window_size[1], step_size):
        for x in range(0, w - window_size[0], step_size):
            yield (x, y, image[y:y + window_size[1], x:x + window_size[0]])

# Load model hasil training
model = YOLO('runs/detect/sawitpro_model/weights/best.pt')

# Membaca gambar besar
image_path = r"D:\backup\download\house-prices-advanced-regression-techniques\SAWITPRO\count\count.jpeg"
image = cv2.imread(image_path)

# Menyimpan hasil deteksi
output_dir = "detected_outputs/"
os.makedirs(output_dir, exist_ok=True)

# Variabel untuk menyimpan hasil deteksi (bounding boxes dan posisi window)
detected_windows = []

# Variabel untuk menghitung jumlah kelapa sawit terdeteksi
total_sawit_detected = 0

# Menjalankan deteksi pada setiap window
for (x, y, window) in sliding_window(image):
    # Simpan setiap window sementara
    temp_file = f"{output_dir}temp_window_{x}_{y}.jpg"
    cv2.imwrite(temp_file, window)
    
    # Deteksi pada window
    results = model.predict(source=temp_file, save=True)
    
    # Hasil deteksi ada dalam list, jadi kita ambil elemen pertama
    result = results[0]  # Ambil objek pertama dari list hasil deteksi
    
    # Menyimpan koordinat bounding box dan posisi window
    detected_windows.append((result.boxes.xyxy, (x, y)))

    # Hitung jumlah kelapa sawit terdeteksi di setiap window
    total_sawit_detected += len(result.boxes)

# Tampilkan jumlah kelapa sawit terdeteksi
print(f"Total kelapa sawit yang terdeteksi di gambar utuh: {total_sawit_detected}")
```

```python
# 6. Combine Image
import cv2
import numpy as np
from pathlib import Path

def combine_image_pieces_with_text(original_image_path, cropped_images_folder, total_detections, window_size=(512, 512), step_size=256):
    """
    Menggabungkan potongan gambar ke dalam gambar utuh dan menambahkan keterangan teks.
    """
    # Membaca gambar asli untuk dimensi
    original_image = cv2.imread(original_image_path)
    if original_image is None:
        print("Error: Gambar asli tidak ditemukan atau tidak dapat dibaca.")
        return None

    original_h, original_w, _ = original_image.shape
    print(f"Dimensi gambar asli: {original_h}x{original_w}")

    # Gambar kosong untuk penggabungan
    combined_image = np.zeros_like(original_image)

    # Membaca semua potongan gambar dari folder
    for cropped_image_path in Path(cropped_images_folder).rglob('*.jpg'):
        print(f"Processing file: {cropped_image_path.stem}")
        
        # Hapus bagian 'temp_window_' dari nama file
        split_name = cropped_image_path.stem.replace("temp_window_", "").split("_")

        # Cek apakah nama file memiliki format yang valid
        if len(split_name) != 2:
            print(f"Skipping file due to invalid name format: {cropped_image_path.stem}")
            continue

        # Ambil offset dari nama file
        try:
            x_offset, y_offset = map(int, split_name)
        except ValueError as e:
            print(f"Error processing file {cropped_image_path.stem}: {e}")
            continue

        # Membaca potongan gambar
        cropped_image = cv2.imread(str(cropped_image_path))
        if cropped_image is None:
            print(f"Error: Tidak dapat membaca gambar {cropped_image_path}")
            continue

        # Menempatkan potongan gambar di posisi yang sesuai
        combined_image[y_offset:y_offset + window_size[1], x_offset:x_offset + window_size[0]] = cropped_image

    # Tambahkan keterangan jumlah kelapa sawit ke gambar
    text = f"Total Sawit: {total_detections}"
    font_scale = 20
    color = (0, 0, 0) 
    thickness = 30
    position = (50, 1000)  
    cv2.putText(combined_image, text, position, cv2.FONT_HERSHEY_SIMPLEX, font_scale, color, thickness)

    return combined_image

def crop_black_borders(image):
    # Konversi gambar ke grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    # Buat mask untuk area non-hitam
    _, mask = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
    
    # Cari koordinat bounding box dari area non-hitam
    coords = cv2.findNonZero(mask)  # Mendapatkan semua koordinat non-hitam
    x, y, w, h = cv2.boundingRect(coords)  # Buat bounding box dari koordinat
    
    # Potong gambar menggunakan bounding box
    cropped_image = image[y:y+h, x:x+w]
    return cropped_image

# Path ke gambar asli
original_image_path = r"D:\backup\download\house-prices-advanced-regression-techniques\SAWITPRO\count\count.jpeg"

# Folder potongan gambar
cropped_images_folder = "runs/detect/predict2"

# Total deteksi kelapa sawit
total_detections = 3889  # Sesuaikan dengan hasil deteksi Anda

# Gabungkan potongan gambar dan tambahkan teks
combined_image_with_text = combine_image_pieces_with_text(
    original_image_path,
    cropped_images_folder,
    total_detections
)

if combined_image_with_text is not None:
    # Potong area hitam dari gambar gabungan
    cropped_image = crop_black_borders(combined_image_with_text)
    
    # Tampilkan gambar yang sudah dipotong
    cv2.imshow("Gambar Gabungan Final", cropped_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Simpan gambar yang sudah dipotong
    output_path = "gambar final.jpg"
    cv2.imwrite(output_path, cropped_image)
    print(f"Gambar gabungan disimpan di: {output_path}")
```
