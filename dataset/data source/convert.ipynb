```python
import tensorflow as tf
import os
import xml.etree.ElementTree as ET

def xml_to_tfrecord(output_path, image_dir, annotations_dir):
    with tf.io.TFRecordWriter(output_path) as writer:
        for xml_file in os.listdir(annotations_dir):
            if xml_file.endswith('.xml'):
                tree = ET.parse(os.path.join(annotations_dir, xml_file))
                root = tree.getroot()
                filename = root.find('filename').text
                image_path = os.path.join(image_dir, filename)
                with tf.io.gfile.GFile(image_path, 'rb') as fid:
                    encoded_image = fid.read()

                # Buat contoh TFRecord
                tf_example = serialize_example(encoded_image, root)
                writer.write(tf_example)

def serialize_example(encoded_image, root):
    # Tambahkan atribut TFRecord seperti bounding box dan label
    features = {
        "image/encoded": tf.train.Feature(bytes_list=tf.train.BytesList(value=[encoded_image])),
        # Tambahkan bounding box dan label lainnya di sini
    }
    return tf.train.Example(features=tf.train.Features(feature=features)).SerializeToString()

xml_to_tfrecord(
    output_path='train.tfrecord',
    image_dir='D:/backup/download/house-prices-advanced-regression-techniques/SAWIT PRO/palm tree.v1i.voc/train/',
    annotations_dir='D:/backup/download/house-prices-advanced-regression-techniques/SAWIT PRO/palm tree.v1i.voc/train/'
)
```
